```python
from bs4 import BeautifulSoup
import requests
import certifi
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
URL = "https://www.ttilgb.com/uiAgs03Action/openScreen.do"
content = requests.get(URL, verify=False)
```

```python
soup = BeautifulSoup(content.text, 'html.parser')
```

```python
row = soup.find('table') # Extract and return first occurrence of tr
print(row)            # Print row with HTML formatting
print("=========Text Result==========")
print(row.get_text()) # Print row as text
```
